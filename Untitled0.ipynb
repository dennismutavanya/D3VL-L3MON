```python
import matplotlib.pyplot as zu
x1 = [2, 3, 4]
y1 = [5, 5, 5]
x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]
zu.scatter(x1, y1, label="gw")
zu.scatter(x2, y2, label="geo", marker='1', color='r')
zu.scatter(x2, y3, label="wainaina", marker='^', color='m')
zu.title('Scatter Plot')
zu.legend()
zu.show()
```
